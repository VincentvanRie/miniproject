```python
from main import prepareCircuit, cost_function, cost_functionNoisy, createHubbardHamiltonian, S
import sympy
import numpy
from scipy.optimize import minimize
import openfermion
```

### Print initial circuit

```python
params = [[sympy.Symbol(f'θ_U{i}'), sympy.Symbol(f'θ_v{i}'), sympy.Symbol(f'θ_h{i}')] for i in range(S)]
params = numpy.array(params).flatten()

print(prepareCircuit(params))
```

### Calculate Noiseless energy

```python
optimisedEnergies = {}

for i in range(3, 12, 2):
    energies = []
    for _ in range(10):
        init_guess = [[0.5, 0.5, 0.5] for _ in range(i)]
        init_guess = numpy.array(init_guess).flatten()

        opt_result = minimize(cost_function, init_guess, method='cobyla')
        energies.append(opt_result.fun)
    optimisedEnergies[i] = energies

hamiltonian = sum(createHubbardHamiltonian().values())

for i in optimisedEnergies:
    print(f'Optimised energy for S={i}: {numpy.mean(optimisedEnergies[i])} ± {numpy.std(optimisedEnergies[i])}')
# print('Optimised parameter: \n', opt_result.x)
# print('Optimised VQE result:', opt_result.fun)
print('Target Hamiltonian eigenvalues:', openfermion.eigenspectrum(hamiltonian)[:3])
```

### Calculate Noiseful energy

```python
optimisedNoisyEnergies = {}

for i in range(3, 12, 2):
    energiesNoisy = []
    for _ in range(6):
        init_guess = [[0.5, 0.5, 0.5] for _ in range(i)]
        init_guess = numpy.array(init_guess).flatten()

        opt_result = minimize(cost_functionNoisy, init_guess, method='cobyla')
        energiesNoisy.append(opt_result.fun)
    optimisedNoisyEnergies[i] = energiesNoisy

hamiltonian = sum(createHubbardHamiltonian().values())

for i in optimisedNoisyEnergies:
    print(f'Optimised energy for S={i}: {numpy.mean(optimisedNoisyEnergies[i])} ± {numpy.std(optimisedNoisyEnergies[i])}')
# print('Optimised parameter: \n', opt_result.x)
# print('Optimised VQE result:', opt_result.fun)
print('Target Hamiltonian eigenvalues:', openfermion.eigenspectrum(hamiltonian)[:3])
```
